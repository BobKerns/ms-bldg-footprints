```python
dom = lxml.html.fromstring(html)
data = {}
for link in dom.xpath('//a/@href'): # select the url in href for all a tags(links)
    
    start = link.rfind('/') + 1
    stop = link.find('.zip')
    state = link[start:stop]
    
    data[state] = link
```

```python
df = pd.DataFrame(data, index=['url']).T
df.index.name='name'
df = df.reset_index()
df['name'] = df['name'].str.replace(' ', '').str.lower()
```

```python
with open('data/states_by_fips.json') as data_file:    
    fips_to_states = json.load(data_file)
states_to_fips = {v['name'].replace(' ', '').lower() : {'fips':k, 'abbrev':v['abbreviation']} for k, v in fips_to_states.items()}
```

```python
def lookup(state, attr):
    
    if state in states_to_fips:
        return states_to_fips[state][attr]
    else:
        print('ERROR', state)
    
df['state'] = df['name'].apply(lookup, attr='abbrev')
df['fips'] = df['name'].apply(lookup, attr='fips')
df['name'] = df['fips'].map(lambda x: fips_to_states[x]['name'])
df = df.reindex(columns=df.columns.sort_values())
```

```python
import os
from itertools import islice

from config import extract_files_folder, split_files_folder

max_chunk_size = 1000000 #a million lines each
```

```python
def split_file(filepath, max_chunk_size, split_files_folder):
    
    counter = 0
    
    # standard geojson header
    header = ['{\n',
              '  "type":"FeatureCollection",\n',
              '  "features":\n',
              '  [\n']
    
    # standard geojson footer
    footer = ['  ]\n',
              '}\n']
    
    # open the input file to split into chunks
    with open(filepath, 'r') as f_input:
        
        # one loop for each chunk
        while True:
            
            # slice out next chunk of lines from the file
            chunk = list(islice(f_input, max_chunk_size))
            
            # if we're out of lines, break out of the loop, we're done
            if not chunk:
                break
            
            else:
                
                # how many lines in this chunk?
                chunk_size = len(chunk)

                # if the first line in the chunk is a feature
                if chunk[0].startswith('    {"type":"Feature"'):
                    # prepend standard geojson header
                    chunk = header + chunk

                # if the last line in the chunk is a feature
                if chunk[-1].startswith('    {"type":"Feature"'):
                    # append standard geojson footer
                    chunk = chunk + footer
                
                # output path to save this chunk
                name = filename[: filename.rfind('.')]
                ext = filename[filename.rfind('.') + 1 :]
                filename_output = '{}-{:03d}.{}'.format(name, counter, ext)
                filepath_output = '{}/{}'.format(split_files_folder, filename_output)
                
                # write the chunk to file
                with open(filepath_output, 'w') as f_output:
                    f_output.writelines(chunk)

                print(filename_output, chunk_size)
                counter += 1
```

```python
# for each geojson file, split it into smaller chunks
for filename in os.listdir(extract_files_folder):
    filepath = '{}/{}'.format(extract_files_folder, filename)
    split_file(filepath, max_chunk_size, split_files_folder)
```
